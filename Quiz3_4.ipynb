```python
import pandas as pd
df = pd.read_csv('/content/Earthquakes_database.csv')
df
```

```python
df.drop_duplicates(inplace=True)
df
#ერთნაირი სტრიქონების წაშლა
```

```python
df.Date = df.Date.str.replace('/','.')
df
#dateში სლეშის შეცვლა წერტილით
```

```python
import numpy as np
df = df.replace('Nan','')
df = df.replace('', np.nan)
df
```

```python
df.Magnitude.value_counts()
#Magnitudeში ვერსიების რაოდენობა
```

```python
bins = [0, 50, 100, 400, 700]
labels = ['Shallow', 'Intermediate', 'Deep', 'Very Deep']
df['Category'] = pd.cut(df.Depth, bins=bins, labels=labels)
df
```

```python
pivot_df = pd.pivot_table(df, values=['Magnitude', 'Depth'], index='Type', columns='Date', aggfunc='mean')
pivot_df
#მომხდარი მიწისძვრების სიღრმე და მაგნიტუდა
```

```python
df.columns
```

```python
melted_data = pd.melt(df, id_vars=['Date', 'Time','Type'],
                      value_vars=['Depth'],
                      var_name='Depth',value_name='Depth_km')
melted_data
#დროის მიხედვით რა სიღრმისა და როგოტი მიწისზსბრა იყო
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(df['Magnitude'], bins=20, kde=False)
plt.xlabel('Magnitudes')
plt.ylabel('Frequency')
plt.show()

#მაგნიტუდების მიხედვით მიწისძვრების სიხშირე
```

```python
type_counts = df['Type'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(type_counts, labels=type_counts.index, autopct='%.1f%%', startangle=90)
plt.legend(labels=type_counts.index, loc='upper right')
plt.show()
# ტიპების სიხშირე
```

```python
categories_counts = df.Category.value_counts()
categories_counts

plt.figure(figsize=(10, 6))
bars = plt.bar(categories_counts.index, categories_counts.values)

plt.xlabel('Depth Categories')
plt.ylabel('Count')
plt.title('Distribution of Earthquake Depth')

plt.legend(bars, type_counts.index, loc='upper right')

plt.show()
# სიღრმის სიხშირე
```

```python

df[['Day', 'Month', 'Year']] = df['Date'].str.split('.', expand=True)
df
#სვეტის 3 ნაწილად გაყოფა
```

```python
df.Depth.value_counts()
```

```python
years = df.Year
depths = df.Depth

plt.figure(figsize=(10, 6))
plt.scatter(years, depths, color='blue', marker='o', s=50, alpha=0.75)

plt.xlabel('Years')
plt.ylabel('Depth (km)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
df.Date = df.Date.astype(str)
df.Time = df.Time.astype(str)

df['Datetime'] = pd.to_datetime(df.Date + ' ' + df.Time, errors='coerce')
df
```

```python

df.set_index('Datetime', inplace=True)

df = df[df.index.notnull()]

df.Year = df.index.year
df.Month = df.index.month
df.Day = df.index.day
df.Weekday = df.index.dayofweek

average_magnitude_per_month = df.resample('M')['Magnitude'].mean()
print(average_magnitude_per_month)
#ყოველთვიურად საშუალო მაგნიტუდის მაჩვენებელი
```
